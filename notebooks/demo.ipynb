```gap
D := PetersenGraph();
```

```gap
FrancyDigraph := function(D)
    local graph, nodes, v, w, canvas;
    graph := Graph(GraphType.DIRECTED);
    nodes := [];
    for v in DigraphVertices(D) do
        nodes[v] := Shape(ShapeType.SQUARE, String(v));
        SetColor(nodes[v], "#FFFFFF");    
        Add(graph, nodes[v]);
    od;
    for v in DigraphVertices(D) do
        for w in OutNeighboursOfVertex(D, v) do
            Add(graph, Link(nodes[v], nodes[w]));
        od;
    od;
    canvas := Canvas("A digraph");
    Add(canvas, graph);
    return canvas;
end;
```

```gap
Draw(FrancyDigraph(JohnsonDigraph(5, 2)));
```
